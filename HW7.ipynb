__3.__ Решить систему линейных уравнений методом $LU$-разложения

$$\begin{cases}
2x_{1}+x_{2}+3x_{3}=1 \\
11x_{1}+7x_{2}+5x_{3}=-6 \\
9x_{1}+8x_{2}+4x_{3}=-5
\end{cases}$$

Матрица коэффициентов имеет вид

$$\begin{pmatrix}
2 & 1 & 3 \\ 
11 & 7 & 5\\ 
9 & 8 & 4
\end{pmatrix}.$$

Домножим первую строку на $\frac{11}{2}$ и вычтем ее из второй строки. При этом запишем в матрицу $L$ на место элемента $l_{21}$ значение $\frac{11}{2}$. Затем нам нужно вычесть из третьей строки первую, домноженную на $\frac{9}{2}$. Значит, в матрицу $L$ запишем $l_{31}=\frac{9}{2}$. 

Матрица коэффициентов будет иметь вид

$$\begin{pmatrix}
2 & 1 & 3 \\ 
0 & \frac{3}{2} & -\frac{23}{2}\\ 
0 & \frac{7}{2} & -\frac{19}{2}
\end{pmatrix},$$

а матрица $L$ —

$$\begin{pmatrix}
1 & 0 & 0 \\ 
\frac{11}{2} & 1 & 0 \\ 
\frac{9}{2} & l_{32} & 1
\end{pmatrix}.$$

Далее умножим вторую строку на $\frac{7}{3}$ и вычтем ее из третьей. При этом в матрицу $L$ на место $l_{32}$ запишем $\frac{7}{3}$.

В итоге получим

$$U=\begin{pmatrix}
2 & 1 & 3 \\ 
0 & \frac{3}{2} & -\frac{23}{2}\\ 
0 & 0 & \frac{52}{3}
\end{pmatrix},$$

$$L=\begin{pmatrix}
1 & 0 & 0 \\ 
\frac{11}{2} & 1 & 0 \\ 
\frac{9}{2} & \frac{7}{3} & 1
\end{pmatrix}.$$

Решим теперь систему 

$$Ly=b:$$

$$\begin{cases}
y_{1}=1, \\
\frac{11}{2}y_{1}+y_{2}=-6, \\
\frac{9}{2}y_{1}+\frac{7}{3}y_{2}+y_{3}=-5.
\end{cases}$$

$$y_{1}=1,$$
$$y_{2}=-\frac{23}{2},$$
$$y_{3}=\frac{52}{3}.$$

И затем систему

$$Ux=y:$$

$$\begin{cases}
2x_{1}+x_{2}+3x_{3}=1, \\
\frac{3}{2}x_{2}-\frac{23}{2}x_{3}=-\frac{23}{2}, \\
\frac{52}{3}x_{3}=\frac{52}{3}.
\end{cases}$$

$$x_{3}=1,$$
$$x_{2}=0,$$
$$x_{1}=-1.$$

__4.__ Решить систему линейных уравнений методом Холецкого

$$\begin{cases}
81x_{1}-45x_{2}+45x_{3}=531 \\
-45x_{1}+50x_{2}-15x_{3}=-460 \\
45x_{1}-15x_{2}+38x_{3}=193
\end{cases}$$

Матрица коэффициентов имеет вид

$$\begin{pmatrix}
81 & -45 & 45 \\ 
-45 & 50 & -15\\ 
45 & -15 & 38
\end{pmatrix}.$$

```python
import numpy as np
```

```python
a = np.array([[81, -45, 45], [-45, 50, -15], [45, -15, 38]])
w, v = np.linalg.eig(a)
 
print(f'Матрица A:\n{a}')
print(f'Определитель:\n{np.linalg.det(a):.0f}')
print(f'Собственные значения:\n{w}')
print(f'Собственные векторы:\n{v}')
```

Так как матрица коэффициентов невырожденная, симметричная и положительно определенная, то можно применить метод Холецкого.

Произведем разложение на $LL^{T}$:

$$l_{11}=\sqrt{a_{11}}=\sqrt{81}=9,$$
$$l_{21}=\frac{a_{21}}{l_{11}}=-\frac{45}{9}=-5,$$
$$l_{31}=\frac{a_{31}}{l_{11}}=\frac{45}{9}=5,$$

$$l_{22}=\sqrt{a_{22}-l_{21}^{2}}=\sqrt{50 - 25}=\sqrt{25}=5,$$
$$l_{32}=\frac{1}{l_{22}}\left ( a_{32}-l_{21}l_{31} \right)=\frac{1}{5}\left ( -15+5\cdot5 \right )=2,$$

$$l_{33}=\sqrt{a_{33}-l_{31}^{2}-l_{32}^{2}}=\sqrt{38-25-4}=3.$$

Получили матрицу 

$$L = \begin{pmatrix}
9 & 0 & 0 \\ 
-5 & 5 & 0 \\ 
5 & 2 & 3
\end{pmatrix}, 
\; \; 
L^{T} = \begin{pmatrix}
9 & -5 & 5 \\ 
0 & 5 & 2 \\ 
0 & 0 & 3
\end{pmatrix}.$$

Решим систему $Ly=b:$

$$\begin{cases}
9y_{1}=531, \\
-5y_{1}+5y_{2}=-460, \\
5y_{1}+2y_{2}+3y_{3}=193.
\end{cases}$$

$$y_{1} = 59,$$

$$y_{2} = -33,$$

$$y_{3} = -12.$$

И решим систему $L^{T}x=y:$

$$\begin{cases}
9x_{1}-5x_{2}+5x_{3}=59, \\
5x_{2}+2x_{3}=-33, \\
3x_{3}=-12.
\end{cases}$$

$$x_{3} = -4,$$

$$x_{2} = -5,$$

$$x_{1} = 6.$$
