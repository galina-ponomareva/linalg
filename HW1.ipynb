__1.__ Исследовать на линейную зависимость:

$$f_{1}(x)=e^{x}, f_{2}(x)=1, f_{3}(x)=x+1, f_{4}(x)=x-e^{x}.$$

Заметим, что $f_3(x) - f_2(x) - f_1(x) = x + 1 - 1 - e^x = x - e^x = f_4(x)$, то есть один из векторов является линейной комбинацией двух остальных. Из этого следует, что векторы линейно зависимы.

__2.__ Исследовать на линейную зависимость:
$$f_{1}(x)=2, f_{2}(x)=x, f_{3}(x)=x^{2}, f_{4}(x)=(x+1)^{2}$$

Заметим, что $f_3(x) + 2f_2(x) + \frac{1}{2}f_1(x) = x^2 + 2x + \frac{1}{2} \cdot 2 = x^2 + 2x + 1 = (x+1)^2 = f_4(x)$, то есть один из векторов является линейной комбинацией двух остальных. Из этого следует, что векторы линейно зависимы.

__3.__ Найти координаты вектора $x = (2, 3, 5)\in \mathbb{R}^{3}$ в базисе $b_{1}=(0, 0, 10)$, $b_{2}=(2, 0, 0)$, $b_{3}=(0, 1, 0)$.

$$x=(2, 3, 5)=(0,0,5)+(2, 0, 0)+(0, 3, 0)=\frac{1}{2}\cdot(0, 0, 10)+1\cdot(2, 0, 0)+3\cdot(0, 1, 0)=\frac{1}{2}b_{1}+1b_{2}+3b_{3},$$
то есть координатами вектора $x$ в базисе $b_{1}=(0, 0, 10)$, $b_{2}=(2, 0, 0)$, $b_{3}=(0, 1, 0)$ являются $0.5$, $1$, $3$.

__4.__ Найти координаты вектора $3x^{2}-2x+2\in\mathbb{R}^{3}[x]$:

а) в базисе $1$, $x$, $x^{2}$;

б) в базисе $x^{2}$, $x-1$, $1$.

а) $$3x^{2}-2x+2=2+(-2x)+3x^2=2\cdot1+(-2)\cdot x+3\cdot x^2,$$

то есть координатами вектора $3x^{2}-2x+2$ в базисе $1$, $x$, $x^{2}$ являются $2$, $-2$, $3$.

б) $$3x^{2}-2x+2=3x^2-2(x-1)=3\cdot x^2+(-2)\cdot(x-1)+0\cdot1,$$

то есть координатами вектора $3x^{2}-2x+2$ в базисе $x^{2}$, $x-1$, $1$ являются $3$, $-2$, $0$.

__5.__ Установить, является ли линейным подпространством:

а) совокупность всех векторов трехмерного пространства, у которых по крайней мере одна из первых двух координат равна нулю;
    
б) все векторы, являющиеся линейными комбинациями данных векторов $\{u_{1}, u_{2}, \ldots, u_{n}\}$.

а) совокупность всех векторов трехмерного пространства, у которых по крайней мере одна из первых двух координат равна нулю - это множество всех векторов вида $(0, a, b)$, $(c, 0, d)$ и $(0, 0, e)$.

Данное множество векторов будет линейным подпространством, если сумма любых двух элементов множества, а также произведение любого элемента на любое действительное число будут принадлежать данному множеству.

$$(0, a1, b1)+(0, a2, b2) = (0, a1+a2, b1+b2)$$

$$(0, a, b)+(c, 0, d) = (c, a, b+d)$$

Векторы вида $(c, a, b+d)$ уже не принадлежат совокупности всех векторов трехмерного пространства, у которых по крайней мере одна из первых двух координат равна нулю, значит как минимум одно из условий не выполнено и данное множество не является подпространством линейного пространства $\mathbb{R}^{3}$.

б) все векторы, являющиеся линейными комбинациями данных векторов $\{u_{1}, u_{2}, \ldots, u_{n}\}$ - множество векторов вида $\sum_{t=1}^{n}\lambda_{t}u_{t}$.

$$\sum_{t=1}^{n}\lambda_{t}u_{t}+\sum_{t=1}^{n}\mu_{t}u_{t} = \sum_{t=1}^{n}(\lambda_{t}u_{t}+\mu_{t}u_{t}) = \sum_{t=1}^{n}(\lambda_{t}+\mu_{t})u_{t}$$

$$\alpha\cdot\sum_{t=1}^{n}\lambda_{t}u_{t} = \sum_{t=1}^{n}(\alpha\cdot\lambda_{t}u_{t}) = \sum_{t=1}^{n}(\alpha\cdot\lambda_{t})u_{t}$$

Полученные векторы также принадлежат множеству всех векторов, являющихся линейными комбинациями данных векторов $\{u_{1}, u_{2}, \ldots, u_{n}\}$, то есть данное множество является подпространством линейного пространства $\mathbb{R}^{3}$.
